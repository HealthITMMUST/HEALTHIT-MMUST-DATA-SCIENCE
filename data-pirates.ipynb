```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
academicDetails = pd.read_csv("dataset/academic_detail.csv")
academic = pd.read_csv("dataset/academic.csv")
fos = pd.read_csv("dataset/field_of_study.csv")
origin = pd.read_csv("dataset/origin.csv")
sof = pd.read_csv("dataset/source_of_fund.csv")
status = pd.read_csv("dataset/status.csv")
```

```python
usSdtntsMean = academic['us_students'].mean()
undergraduateMean = academic['undergraduate'].mean()
gradduateMean = academic['graduate'].mean()
non_degreeMean = academic['non_degree'].mean()
optMean = academic['opt'].mean()

academic.us_students.fillna(usSdtntsMean, inplace=True)
academic.undergraduate.fillna(undergraduateMean, inplace=True)
academic.graduate.fillna(gradduateMean, inplace=True)
academic.non_degree.fillna(non_degreeMean, inplace=True)
academic.opt.fillna(optMean, inplace=True)

academic.isnull().sum()
```

```python
fosStdMean = fos.students.mean()
fos.students.fillna(fosStdMean, inplace=True)

fos.isnull().sum()
```
